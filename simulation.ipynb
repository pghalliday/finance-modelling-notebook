# Simulation

We run simulations to calculate daily time series of financial states.

These states will be associated with entities, such as individuals, corporations, etc.

The states will include such things as cash balances, investments, tax liability, tax deductions, etc.

```python
from datetime import date
from src.financial_simulator.entity.corporation import Corporation, CorporationState
from src.financial_simulator.entity.individual import Individual, IndividualState
from src.financial_simulator.salaries.simple_monthly_salary import SimpleMonthlySalary
from src.financial_simulator.financial_simulator import run_simulation
import plotly.express as px
import pandas as pd

FredSalary = SimpleMonthlySalary(18, 2000, 100, 150)
JaneSalary = SimpleMonthlySalary(18, 2500, 125, 175)

Fred = Individual(salaries={'WidgetsBV': FredSalary})
Jane = Individual(salaries={'WidgetsBV': JaneSalary})

WidgetsBV = Corporation(salaries={'Fred': FredSalary,
                                  'Jane': JaneSalary})

days = run_simulation(start_date=date(2020, 1, 1),
                      end_date=date(2025, 12, 31),
                      entities={'Fred': Fred,
                                'Jane': Jane,
                                'WidgetsBV': WidgetsBV},
                      state={'Fred': IndividualState(cash=10000.0,
                                                     box_1_tax_paid=0.0,
                                                     box_1_taxable_income=0.0),
                             'Jane': IndividualState(cash=15000.0,
                                                     box_1_tax_paid=0.0,
                                                     box_1_taxable_income=0.0),
                             'WidgetsBV': CorporationState(cash=200000.0)
                      })

df = pd.DataFrame([x
                   for xs in [[{'date': day.date,
                                'cash': day.state['Fred'].cash,
                                'entity': 'Fred'},
                               {'date': day.date,
                                'cash': day.state['Jane'].cash,
                                'entity': 'Jane'},
                               {'date': day.date,
                                'cash': day.state['WidgetsBV'].cash,
                                'entity': 'WidgetsBV'},
                               {'date': day.date,
                                'cash': day.state['Fred'].cash +
                                        day.state['Jane'].cash +
                                        day.state['WidgetsBV'].cash,
                                'entity': 'Total'}]
                               for day in days]
                   for x in xs])
fig = px.line(df, x='date', y='cash', color='entity')
fig.to_image()
```
